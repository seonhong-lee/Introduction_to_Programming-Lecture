## break와 continue



*   break문



```python
s = "what are the 10 best-selling products in this shopping mall?"
idx = 0
while idx < len(s):
  if s[idx].isdigit():
    print(idx)
  idx += 1
```

```python
s = "what are the 10 best-selling products in this shopping mall?"
idx = 0
found = False
while idx < len(s):
  if found == False and s[idx].isdigit(): # 첫 번째 숫자를 못찾으면
    print(idx)
    found = True # 첫 번째 숫자를 찾음
  idx += 1
```

```python
s = "what are the 10 best-selling products in this shopping mall?"
idx = 0
found = False
while found == False and idx < len(s):
  if found == False and s[idx].isdigit():
    print(idx)
    found = True

  idx += 1
```

```python
s = "what are the 10 best-selling products in this shopping mall?"
idx = 0
found = False
while found == False and idx < len(s):
  if s[idx].isdigit():
    print(idx)
    found = True

  idx += 1
```

```python
s = "what are the 10 best-selling products in this shopping mall?"
idx = 0
while idx < len(s):
  if s[idx].isdigit():
    print(idx)
    break

  idx += 1
```



*   continue문



```python
count = 0
sum = 0
while count < 10:
  n = int(input("양의 정수를 입력하세요: "))
  if n > 0:
    sum += n
    count += 1
    print("count =", count)

print("합 =", sum)
```

```python
count = 0
sum = 0
while count < 10:
  n = int(input("양의 정수를 입력하세요: "))
  if n <= 0: # 0 또는 음수이면
    continue # 반복문의 처음으로 이동

  sum += n
  count += 1
  print("count =", count)
print("합 =", sum)
```
