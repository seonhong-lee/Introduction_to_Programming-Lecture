## 문자열



*   문자열과 인덱스



```python
# 문자열[인덱스]
```

```python
s = "Hi, everyone.\n"
print(s[0]) # 'H
print(s[-1]) # '\n'
print(s[len(s) - 1]) # '\n'
print(s[-2]) # '.'
print(s[10]) # 'n'
```



*   문자열은 수정 불가



```python
s = "Hi, everyone.\n"
s[0] = 'h'
```



*   문자열의 특정 위치 확인하기



```python
# 문자열.index(검색할_문자열)
# 문자열.find(검색할_문자열)
```

```python
s = "Hello World"
print(s.index("Wor")) # 인덱스는 0번부터 시작하여 탐색한다
```

```python
print(s.index("wor"))
```

```python
print(s.find("Wor"))
```

```python
print(s.find("wor"))
```

```python
print(s.index("W"))
```

```python
print(s.find("W"))
```

```python
print(s.index("o"))
```

```python
print(s.find("o"))
```

```python
# 문자열.index(검색할_문자열, 시작_인덱스, 끝_인덱스)
# 문자열.find(검색할_문자열, 시작_인덱스, 끝_인덱스)
```

```python
print(s.index('o', 0, 4))
```

```python
print(s.index('o', 0, 5))
```

```python
print(s.find('o', 0, 4))
```

```python
print(s.find('o', 0, 5))
```



*   rindex()와 rfind() 명령 활용



```python
print("Bye Bye".rindex("Bye"))
```

```python
print("Bye Bye".rfind("Bye"))
```

```python
print("Bye Bye".rindex("Be"))
```

```python
print("Bye Bye".rfind("Be"))
```

```python
# 문자열.rindex(검색할_문자열, 시작_인덱스, 끝_인덱스)
# 문자열.rfind(검색할_문자열, 시작_인덱스, 끝_인덱스)
```



*   문자열을 일부 추출


```python
# 문자열[시작_인덱스:마지막_인덱스]
# 문자열의 슬라이싱
```

```python
s = "Hello World"
print(s[0:2])
```

```python
print(s[1:4]) # 인덱스 1부터 인덱스 3까지
```

```python
print(s[:-1]) # 처음부터 -2까지, 즉 끝에서 한 문자 빼고 추출
```

```python
print(s[:3]) # 처음부터 2까지
```

```python
print(s[:]) # 처음부터 끝까지
```



*   실습문제6: "Beautiful.Image.png" 파일 이름에서 확장자를 제외한 부분을 화면에 출력한다.
*   요구사항:
*   1) 확장자는 파일 이름의 마지막에 ".확장자" 형태로 주어진다고 가정한다.
*   2) 파일 이름에 "."가 두 개 이상 있을 수 있다(최소 한 개는 존재한다).
*   3) 확장자는 몇 글자인지 정해지지 않았다.

```python
filename = "Beautiful.Image.png"
idx = filename.rindex('.')
print(filename[:idx])
```



*   실습문제7: 사용자로부터 두 개 이상의 단어가 있는 문자열을 입력받고, 첫 번째 단어만 추출해서 출력한다.

*   요구사항:
*   1) 사용자가 반드시 두 개 이상의 단어를 입력한다고 가정한다.
*   2) 단어는 공백 문자로 분리된다(탭문자나 줄바꿈 문자는 검사하지 않는다).
*   3) 문자열의 양끝에 공백 문자가 있을 수 있다(먼저 제거한 후 단어를 추출한다).




```python
s = input("두 개 단어 이상의 문자열을 입력하세요: ")
s = s.strip()
idx = s.index(' ')
print(s[:idx])
```
