## 반복문의 필요성



*   반복문을 사용하지 않는 입력과 합계



```python
# 방법 1
n1 = int(input("정수 한 개를 입력하세요: "))
n2 = int(input("정수 한 개를 입력하세요: "))
n3 = int(input("정수 한 개를 입력하세요: "))

sum = n1 + n2 + n3
print(sum)
```

```python
# 방법 2
sum = 0
n1 = int(input("정수 한 개를 입력하세요: "))
sum += n1
n2 = int(input("정수 한 개를 입력하세요: "))
sum += n2
n3 = int(input("정수 한 개를 입력하세요: "))
sum += n3

print(sum)
```
