## while 반복문



*   반복문은 중복된 코드가 사라지게 하는 것



```python
count = 1
sum = 0
n = int(input("정수 한 개를 입력하세요: ")) # (1)
sum += n                                    # (2)
count += 1                                  # (3)
n = int(input("정수 한 개를 입력하세요: "))
sum += n
count += 1
n = int(input("정수 한 개를 입력하세요: "))
sum += n
count += 1

print(sum)
```

```python
count = 1
sum = 0
while count < 4:
  n = int(input("정수 한 개를 입력하세요: ")) # (1)
  sum += n                                    # (2)
  count += 1                                  # (3)
print(sum)
```



*   while문의 조건식은 True/False에 준하는 정수나 문자열도 사용 가능



```python
print("5개 정수를 입력하면 가장 큰 값을 화면에 출력합니다")
maxValue = 0
count = 5

while count:    # 또는 count > 0
  n = int(input("0보다 큰 정수 한 개를 입력하세요: "))
  if n > maxValue:
    maxValue = n
  count -= 1

print("최대값 = ", maxValue)
```

```python
s = input("출력할 문자열을 입력하세요. 빈 문자열을 입력하면 종료됩니다: ")
while s:
  print(s)
  s = input("출력할 문자열을 입력하세요. 빈 문자열을 입력하면 종료됩니다: ")
```

*   실습문제1: 사용자로부터 숫자 5개를 입력받고, 가장 큰 입력값을 화면에 출력하는 프로그램을 구현한다.

*   요구사항:
*   - 사용자가 입력하는 숫자 5개는 모두 0보다 큰 양수로 가정한다.
*   - 자료의 정렬 알고리즘이 적용된 함수를 사용하지 않고 풀어본다.

```python
print("5개 정수를 입력하면 가장 큰 값을 화면에 출력합니다")
maxValue = 0
count = 1
while count < 6:
  n = int(input("0보다 큰 정수 한 개를 입력하세요: "))
  if n > maxValue:
    maxValue = n
  count += 1

print("최대값= ", maxValue)
```

*   실습문제2: 사용자로부터 1부터 100 사이의 정수를 입력받고 화면에 출력하는 프로그램을 작성한다. 1부터 100 사이 범위 밖의 정수가 입력되면 범위 안의 정수가 입력될 때까지 계속 입력받는다.

*   요구사항:
*   - 사용자가 입력하는 숫자는 음수와 양수를 포함하는 정수로 제한한다.
*   - 1부터 100사이의 정수는 1, 2, ..., 99, 100을 나타낸다.

```python
n = int(input("1~100 사이의 정수를 한 개 입력하세요: "))
while n < 1 or n > 100:                                    # (1)
  n = int(input("1~100 사이의 정수를 한 개 입력하세요: ")) # (2)
print(n)
```
