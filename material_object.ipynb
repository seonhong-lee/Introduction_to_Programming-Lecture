```python
ll = [str, int, list, tuple, type(None), object, type]

for i in dir(object):
  if (type(object.__getattribute__(object, i)) in ll):
    print(i)
```

```python
count = 0
for i in dir(object):
  if (type(object.__getattribute__(object, i)) not in ll):
    print(i, end=", ") # 내장 클래스가 아닌 경우에 출력한다. 함수나 메소드를 출력.
    count += 1
    if count % 5 == 0:
      print()
```
