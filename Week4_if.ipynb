*   조건  | 비교 연산 | 출력 내용
*   5 이상|num >= 5   | 5보다 크거나 같습니다
*   5 미만|num < 5    | 5보다 작습니다

*요구사항: 사용자가 오류 없이 정수만을 입력한 것이라고 가정한다.



```python
num = int(input("정수 한 개를 입력하세요: "))
if num >= 5:
  print("5보다 크거나 같습니다")
if num < 5:
  print("5보다 작습니다")
```



*   실습문제2: 사용자로부터 정수 한 개를 입력받고, 정수값이 10이상, 5이상, 10미만, 5미만이라면, 아래 표에 보이는 것처럼 내용을 화면에 출력한다.
*   번호 |   조건     |    파이썬 비교 연산    |    출력 내용
*      1 |10 이상     |       num >= 10       | 10보다 크거나 같습니다
*      2 |5이상 10미만|     5 <= num < 10     | 5보다 크거나 같지만 10보다 작습니다
*      3 | 5 미만     |        num < 5        | 5보다 작습니다

*   요구사항: 사용자가 오류 없이 정수만을 입력할 것이라고 가정한다.         



```python
num = int(input("정수 한 개를 입력하세요: "))
if num >= 10:
  print("10보다 크거나 같습니다")
else:
  if 5 <= num: # if num >= 5: 라고 써도 됨
    print("5보다 크거나 같지만 10보다 작습니다")
  else:        # if num < 5
    print("5보다 작습니다")
```

*   요구사항
*   - 확장자는 파일 이름의 마지막에 ".확장자" 형태로 주어진다고 가정한다.
*   - 파일 이름에 "."가 없을 수 있다. 확장자가 없다면, 입력된 파일 이름 전체를 출력한다.
*   - 파일 이름에 "."가 두 개 이상 있을 수 있다.
*   - 확장자의 글자 수는 정해지지 않았다.



```python
filename = input("파일 이름을 입력하세요: ")
idx = filename.rfind('.')
if idx < 0:
  idx = len(filename)
print(filename[:idx])
```

*   요구사항
*   - 파일 이름이 ".png"로 끝나는지 확인해야만 한다.
*   - replace() 함수를 사용하지 않고 다른 명령들을 이용해서 문제를 해결한다.



```python
filename = input("파일 이름을 입력하세요: ")
if filename.endswith(".png"):
  print(filename[:-4] + ".jpg")
else:
  print(filename)
```

*   좌석 종류 | 가격
*     VIP     | 150000
*      S      | 110000
*      A      | 90000
*      B      | 70000

*   사용자로부터 구매할 좌석의 종류를 입력받고, 그 좌석에 해당되는 가격을 출력하는 프로그램을 작성한다.

*   요구사항: 사용자가 입력하는 좌석 종류가 "VIP", "S", "A", "B" 중 한 가지가 아니라면, 좌석 종류를 잘못 입력했다고 출력한다.

*   항목 추가



```python
seatType = input ("좌석 종류를 입력하세요: ")
if seatType == "VIP":
  print("티켓 가격은 150000입니다")
elif seatType == "S":
  print("티켓 가격은 110000입니다")
elif seatType == "A":
  print("티켓 가격은 90000입니다")
elif seatType == "B":
  print("티켓 가격은 70000입니다")
else:
  print("좌석 종류를 잘못 입력했습니다")
```



*   실습문제6: 사용자에게 온도와 습도를 정수로 입력받고, 다음 조건에 맞춰 에어컨을 켜는 프로그램을 작성한다.

*   - 온도가 25도 이상이고 습도가 70 이상이면, 에어컨을 냉방 모드로 켠다.
*   - 온도가 25도 이상이지만, 습도가 70미만일 때 제습 모드로 작동시킨다.
*   - 두 조건에 만족되지 않으면 에어컨을 끈다.

*   요구사항:
*   - 사용자가 정수만을 입력할 것이라고 가정한다.
*   - 실제 에어컨을 켜고 끌 수 없으므로, "냉방 모드로 에어컨 작동" 또는 "제습 모드로 에어컨 작동"이라고 출력한다.



```python
temperature = int(input("온도를 입력하세요: "))
humidity = int(input("습도를 입력하세요: "))

if (temperature >= 25) and (humidity >= 70):   # 온도 25도 이상 습도 70 이상
   print("에어컨을 냉방 모드로 켬")
elif (temperature >= 25) and (humidity < 70):  # 온도 25도 이상 습도 70 미만
  print("에어컨을 제습 모드로 켬")
else:
  print("에어컨을 끔")
```
