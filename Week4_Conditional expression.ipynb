## 조건식

```python
# 실내_온도 >= 사용자_지정_온도
```



*   문자열 비교



```python
print("한글" == "한글")
print("한글" != "English")
print("가" < "나")            # 같은 언어 비교시 사전 순
print("English" == "English")
print("English" < "Korean")   # 같은 언어 비교시 사전 순
print("English" < "한글")     # 영문 < 한글
print("1234" < "A")           # 숫자 < 알파벳
```



*   숫자 비교



```python
print(2 == 2.0)
```

```python
print(2 < 2.1)
print(2.1 >= 2.3)
```



*   불린 비교



```python
print(True == True)
print(True == False)
```

```python
print((2 > 3) == (3 > 2)) # False == True이므로 False임
print((3 > 2) == (2 < 3)) # True == True이므로 True임
```

```python
print(int(True))
print(int(False))
print(True > False)
print(True < False)
```

```python
print((3 > 2) > (2 < 3)) # True < True이므로 False
```



*   다른 자료형끼리 비교



```python
print("abc" == (3 > 2)) # 문자열과 불린값 비교
print("abc" == 3.2)     # 문자열과 실수값 비교
print("3.2" == 3.2)     # 문자열과 실수값 비교
print("abc" != (3 > 2)) # 문자열과 불린값 비교
print("abc" != 3.2)     # 문자열과 실수값 비교
print("3.2" != 3.2)     # 문자열과 실수값 비교
```

```python
"abc" > True
```

```python
"123" > 123
```

```python
print(1.1 > True)
print(123 > True)
```
