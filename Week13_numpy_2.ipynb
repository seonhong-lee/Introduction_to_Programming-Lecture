## 연산을 함수로 사용해보기(배열 두개)



*   더하기 add / 빼기 substract / 곱하기 multiply / 나누기 divide



```python
a = np.array([1, 2, 3])
b = np.array([3, 2, 1])

c1 = np.add(a, b)
c2 = np.subtract(a, b)
c3 = np.multiply(a, b)
c4 = np.divide(a, b)

print(c1)
print(c2)
print(c3)
print(c4)
```



*   maximum 최대값 / minimum 최소 값


```python
a = np.array([2, 3, 4])
b = np.array([1, 5, 2])

c1 = np.maximum(a, b)
c2 = np.minimum(a, b)

print(c1)
print(c2)
```

## 원소들의 통계함수(배열 한개)



*   합계 sum / 누적합 cumsum / 곱 prod / 누적곱 cumprod



```python
a = np.array([1, 2, 3, 4, 5])

b1 = np.sum(a)
b2 = np.cumsum(a)
b3 = np.prod(a)
b4 = np.cumprod(a)

print(b1)
print(b2)
print(b3)
print(b4)
```



*   평균 mean / 분산 var / 표준편차 std



```python
b1 = np.mean(a)
b2 = np.std(a)
b3 = np.var(a)

print(b1)
print(b2)
print(b3)
```



*   최대값 max / 최소값 min



```python
b1 = np.max(a)
b2 = np.min(a)

print(b1)
print(b2)
```

## 정렬함수



*   1차원배열 오름차순 / 내림차순 정렬



```python
a = np.array( [ 0.2548745, 0.0548778, 0.11547, 0.2548744, -1.215487 ] )
```

```python
b1 = np.sort(a)
b2 = np.sort(a)[::-1]

print(b1)
print(b2)
```



*   2차원배열 오름차순 / 내림차순 정렬



```python
a = np.array( [ [ 11, 5, 3 ],
                [ 2, 8, 9 ],
                [ 7, 3, 4] ] )
```

```python
b1 = np.sort(a, axis=0)     #행기준 정렬 3행의 모든 요소가 정렬
b2 = np.sort(a, axis=1)     #열기준 정렬 3열의 모든 요소가 정렬
print(b1)
print(b2)
```
