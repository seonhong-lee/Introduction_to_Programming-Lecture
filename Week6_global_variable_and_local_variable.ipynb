
## 전역 변수와 지역 변수



```python
num1 = 3         # (1)
def printNum():
  num2 = 4       # (2)
  print(num1)    # (3)
  print(num2)    # (4)

print(num1)      # (5)
printNum()       # (6)
```



*   전역 변수와 같은 이름의 지역 변수의 유효 범위



```python
num = 3            # (1)
def printNum2():
  num = 5          # (2)
  print(num)       # (3)

print(num)         # (4)
printNum2()        # (5)
print(num)         # (6)
```



*   지역 변수의 유효 범위는 함수 전체



```python
num = 3          # (1)
def printNum3():
  print(num)     # (2)
  num = 5        # (3)
  print(num)

printNum3()      # (4)
```

```python
num = 3          # (1)
def printNum3():
  print(num)     # (2)

printNum3()      # (3)
```

*   함수에서 전역 변수의 값을 변경

```python
num = 0
def add5(n):
  return n + 5

num = add5(num)     # (1)
print(num)          # (2)
```

```python
num = 0            # (1)
def add5(n):
  global num       # (2)
  num = 5          # (3)

add5(num)
print(num)
```

```python
num1 = 0
num2 = 0

def setNum1AndNum2():
  global num1, num2 # 콤마로 분리해서 나열
  num1 = 5
  num2 = 7

setNum1AndNum2()
print(num1)         # (1)
print(num2)         # (2)
```
