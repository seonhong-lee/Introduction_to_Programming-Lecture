## 예외 처리

## try ... except 구문

```python
print("예외 처리 확인 프로그램")
try:
  s = input("정수를 입력하세요: ")
  n = int(s)
  print(f"사용자가 입력한 정수: {n}")
except:
  print(f"정수로 변환할 수 없는 값을 입력하셨습니다. 입력값: {s}")

print("프로그램 종료")
```

## 여러 오류에 대응하는 예외 처리 코드 작성



*   모든 오류에 대해서 한꺼번에 대응



```python
try:
  lst = [1, 2, 3]
  idx = int(input("변경할 요소의 위치(인덱스)를 입력하세요: "))
  n = int(input("새로운 값을 입력하세요: "))
  lst[idx] = n    # idx 위치의 요소 값을 n으로 변경
  print(lst)
except:           # 모든 오류를 한 곳에서 대응
  print("오류 발생, 인덱스 범위를 벗어났거나 정수가 아닌 문자열을 입력했음")
```



*   각 오류에 대해서 따로 대응



```python
try:
  lst = [1, 2, 3]
  idx = int(input("변경할 요소의 위치(인덱스)를 입력하세요: "))
  n = int(input("새로운 값을 입력하세요: "))
  lst[idx] = n     # idx 위치의 요소 값을 n으로 변경
  print(lst)

except ValueError: # ValueError에 대응
  print("오류 발생, 정수가 아닌 문자열을 입력했음")
except IndexError: # IndexError에 대응
  print("오류 발생, 인덱스 범위를 벗어남")
```
