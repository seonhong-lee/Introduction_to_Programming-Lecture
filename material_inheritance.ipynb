```python
class Polygon:                # 다각형 클래스 정의.
  def __init__(self, no_of_sides): # 초기화 함수로 객체의 속성 추가.
    self.n = no_of_sides           # 변의 개수
    self.x_sides = [0 for i in range(no_of_sides)] # x축 좌표
    self.y_sides = [0 for i in range(no_of_sides)] # y축 좌표

  def inputXSides(self):    # x축 좌표 입력.
    self.x_sides = [float(input("Enter X 좌표 " + str(i+1) + " : ")) for i in range(self.n)]

  def inputYSides(self):    # y축 좌표 입력.
    self.y_sides = [float(input("Enter Y 좌표 " + str(i+1) + " : ")) for i in range(self.n)]

  def dispSides(self):    # 좌표 출력.
    for i in range(self.n):
      print("X 좌표 ", i+1, "is", self.x_sides[i])
      print("Y 좌표 ", i+1, "is", self.y_sides[i])

  def polygonArea(self):     # 다각형 면적 구하기.
    X = self.x_sides
    Y = self.y_sides

    area = 0
    j = self.n - 1           # 좌표를 비교해서 길이를 확인하고 면적을 구한다.
    for i in range(self.n):
      area = area + (X[j] + X[i]) * (Y[j] - Y[i])
      j = i
    return abs(area/2) # 좌표에 의한 계산법 A = 1/2{(x1+x2)(y2-y1)+(x2+x3)(y3-y2)-(x1+x4)(y4-y1)-(x4+x3)(y3-y4)}
```

```python
class Triangle(Polygon):            # 다각형을 상속한 삼각형 클래스 정의
  def __init__(self, no_of_sides):  # 초기화 함수에 변을 3으로 고정
    if no_of_sides == 3:
      super().__init__(no_of_sides)
    else:
      raise TypeError("변의 길이가 다릅니다.")
```

```python
t = Triangle(3)    # 삼각형 객체 생성
```

```python
t.inputXSides()
```

```python
t.inputYSides()
```

```python
t.polygonArea()  # 면적을 구한다.
```

```python
t.dispSides()
```

```python
class Ractangle(Polygon):             # 다각형을 사각형 클래스로 정의.
  def __init__(self, no_of_sides):
    if no_of_sides == 4:
      super().__init__(no_of_sides)   # 4개의 변을 처리할 수 있다.
    else:
      raise TypeError("변이 길이가 다릅니다.")
```

```python
class Person(object):      # Person 클래스 정의.
  def __init__(self, name, age, nation): # 초기화 함수에 3개의 매개변수를 넣는다.
    print("Person 초기화 사용")
    self.name = name
    self.age = age
    self.nation = nation

  def say(self, message): # say 함수는 메세지를 받고 그대로 반환.
    return message
```

```python
class GrandFather(Person):    # GrandFather 클래스 정의.
  pass
```

```python
class GrandMother(Person):    # GrandMother 클래스 정의
  pass
```

```python
class Father(GrandFather, GrandMother): # GrandFather와 GrandMother 클래스를 상속한 Father 클래스 생성
  pass
```

```python
class Animal: # Animal 클래스에 하나의 함수만 정의.
  def say(self):
    print("Animal 호출")
    return self._say

class Species: # Species 클래스에 하나의 함수만 정의.
  def say(self):
    print("Species 호출")
    return self._say
```

```python
class Dog(Animal, Species): # 다중 상속할 때 Animal, Species 순서로 상속.
  def __init__(self, bark): # 초기화 함수를 정의해서 하나의 속성 생성.
    self._say = bark        # 2개의 상속 클래스에 있는 say를 오바라이딩(재정의) 한다.
  def say(self):            # 상위 클래스를 연결해서 say 메소드 실행.
    return super().say()
```

```python
class Bird(Species, Animal): # 다중 상속할 때 Species, Animal 순서로 상속.
  def __init__(self, sing):
    self._say = sing
  def say(self):             # 2개의 상속 클래스에 있는 say를 오버라이딩(재정의)한다.
    return super().say()     # 상위 클래스를 연결해서 say 메소드 실행.
```

```python
d = Dog("멍멍멍")   # 객체 생성.
d.say()             # 메소드 호출
```

```python
s = Bird("짹짹짹") # 객체 생성.
s.say()            # 메소드 호출
```

```python
super(Dog, d) # super 클래스에 Dog 클래스와 객체를 넣으면 다음에 정의된 클래스 참조
```

```python
Dog.mro()
```
