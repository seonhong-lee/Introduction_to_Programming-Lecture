##재귀 호출





*   팩토리얼 함수 구현





```python
def factorial(n):
  if n == 0:
    return 1
  return n * factorial(n - 1)


print(factorial(0))
print(factorial(2))
print(factorial(10))
```



*   최대 공약수 함수 구현



```python
def gcd(m, n):
  if m == n:              # 재귀 호출 탈출 조건
    return m
  elif m > n:
    return gcd(m - n, n)
  else:                   # m < n
    return gcd(m, n - m)

print(gcd(12, 18))
print(gcd(12, 8))
```

*   실습문제5: 인자로 정수 한 개(매개변수 이름은 n)를 입력받고, 1 부터 n까지의 합을 구해서 반환하는 함수를 재귀 호출을 이용해서 구현한다.

*   요구사항:
-  인자로 전달되는 n은 1이상의 정수이다.

```python
def sum(n):
  if n == 1:
    return 1
  return n + sum(n-1)

print(sum(1))
print(sum(10))
print(sum(100))
```
