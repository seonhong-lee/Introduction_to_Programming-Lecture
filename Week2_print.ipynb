

*   end 옵션이 추가된 print() 명령어



```python
print("오늘은 프로그래밍 입문 수업이", end="있는날이다")
```

```python
print("이선홍", end=" ")
```

```python
print("이선홍", end=" (끝)")
```

```python
print("이선홍", end= " (끝)\n")
```

```python
print("이선홍", end= "\n은 오늘 강의를 한다")
```

```python
print("이", "선홍")
```



*   sep 옵션이 추가된 print() 명령어



```python
print("홍", "길동", sep="\n")
```

```python
print("성: 홍", "길동", sep=" 이름: ")
```

```python
print("홍", "길", "동", sep="\n")
```

```python
print("원주율 값", "3.1415", sep= ": ", end="92\n입니다")
```



*   실습문제1: 본인의 이름과 생년월일을 셀 결과에 출력한다.
*   요구사항: 생년월일은 년도-월-일 형태로 출력한다. 예를 들어 생일이 2000년 1월 1일이라고 가정하면 2000-1-1을 출력한다.




```python
print("이선홍")
print("2000-1-1")
```

```python
print("이선홍")
print("2000","1","1", sep="-")
```



*   실습문제2: 본인의 이름과 생년월일을 출력하되, 날짜를 "2000년 1월 1일" 형태로 출력한다.

*   요구사항: print() 명령의 콤마를 사용해서 년/월/일을 분리한다.



```python
print("이선홍")
print("2000년", "1월", "1일")
```



*   실습문제3: 본인의 이름과 생년 월일을 출력하되, 날짜를 "2000/1/1" 형태로 출력한다.
*   요구사항:
1) 년, 월, 일에 해당되는 2000, 1, 1은 문자열이 아닌 숫자 형태로 출력한다.
2) 생년월일은 이름과 같은 줄에 출력한다 (단 print() 명령은 분리해서 사용한다).



```python
print("이선홍", end=" ")
print(2000, 1, 1, sep='/')
```
