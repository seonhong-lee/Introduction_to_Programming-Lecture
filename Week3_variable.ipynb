

*   실습문제 1: 커피 원두 100g의 가격이 10,000원이다. 커피 원두는 최소 200g부터 구매할 수 있고, 100g 단위로 증량해서 구매할 수가 있다. 200g, 300g, 400g을 구매할 때의 가격을 화면에 출력한다.
*   요구사항:

*    1) 변수에 커피 원두 100g의 가격을 지정한 후, 산술 연산을 통해 200, 300, 400g의 가격을 산출해서 출력한다.
*    2) 화면 출력 내용은 다음 형태로 처리한다.

*    커피 원두 200g 가격: 20000원
*    커피 원두 300g 가격: 30000원
*    커피 원두 400g 가격: 40000원




```python
coffee100 = 10000
print("커피 원두 200g 가격:", 200 // 100 * coffee100)
print("커피 원두 300g 가격:", 300 // 100 * coffee100)
print("커피 원두 400g 가격:", 400 // 100 * coffee100)
```



*   실습문제2: 커피 원두 100g의 가격이 10000원이다. 커피 원두는 최소 200g부터 구매할 수 있고, 100g 단위로 증량해서 구매할 수 있다. 200g, 300g, 400g을 구매할 때의 가격을 화면에 출력한다.
*   요구사항: 변수에 커피 원두 100g의 가격을 저장한 후, 복합 연산을 통해 100g의 단가를 차례대로 더해서 200, 300, 400g의 가격을 출력한다.



```python
coffee100 = 10000 # 100g당 단가
coffeePrice = coffee100 # 100g의 커피 원두 가격
coffeePrice += coffee100 # 커피 100g 가격에 100g 단가를 더해 200g의 가격을 계산
print("커피 원두 200g 가격:", coffeePrice)
coffeePrice += coffee100 # 커피 200g 가격에 100g 단가를 더해 300g의 가격을 계산
print("커피 원두 300g 가격:", coffeePrice)
coffeePrice += coffee100 # 커피 300g 가격에 100g 단가를 더해 400g의 가격을 계산
print("커피 원두 400g 가격:", coffeePrice)
```

*   요구사항:
*   1) 커피 무게 (200g, 300g)을 weight라는 변수에 저장한 후 산술 연산식과 print() 명령문에서 사용한다.
*   2) 커피 무게와 가격을 형식 지정자를 이용해서 치환한다.



```python
coffee100 = 10000
weight = 200
print("커피 원두 %dg 가격: %d" % (weight, weight // 100 * coffee100))
weight = 300
print("커피 원두 %dg 가격: %d" % (weight, weight // 100 * coffee100))
```

*    실습문제5: 커피 원두 100g의 가격이 10,000원이다. 커피 원두는 최소 200g부터 구매할 수 있고, 100g 단위로 증량해서 구매할 수가 있다. 200g, 300g을 구매할 때의 가격을 화면에 출력한다.
*   요구사항: 변수 weight를 f-문자열을 이용해서 출력한다.

```python
coffee100 = 10000
weight = 200
print(f"커피 원두 {weight}g 가격: {weight // 100 * coffee100}")
weight = 300
print(f"커피 원두 {weight}g 가격: {weight // 100 * coffee100}")
```
