## for 반복문



*   for 반복문과 문자열



```python
s = "hello"
for ch in s:  # 변수에 저장하지 않고 for ch in "hello". 라고 쓸 수도 있음
  print(ch)
```

```python
for i in range(3):
  print(i, end=' ')
print("")
```

```python
for i in range(1, 5):
  print(i, end=' ')
print("")
```

```python
for i in range(1, 7, 2):
  print(i, end=' ')
print("")
```

```python
for i in range(-5, -2):
  print(i, end=' ')
print("")
```

```python
for i in range(-5, 3, 2):
  print(i, end=' ')
print("")
```

```python
for i in range(7, 1, -3):
  print(i, end=' ')
print("")
```

```python
for i in range(7,-3, -2):
  print(i, end=' ')
print("")
```



*   for문을 이용해서 1부터 10까지 합 구하기



```python
sum = 0
for n in range(1, 11): # n변수값을 1~10까지 변경해서 저장
  sum += n

print(sum)
```

```python
s1 = "simple"
s2 = "python is a simple language to learn"

for ch1 in s1:
  count = 0
  for ch2 in s2:
    if ch1 == ch2:
      count += 1
  print(f"{ch1}:{count}")
```

*   실습문제3: 정수의 약수들을 화면에 출력하는 프로그램을 작성한다. 12와 16의 약수들을 출력해 본다.

*   요구사항:
*   - 정수를 한 개 인자로 전달받고,약수를 화면에 출력한다.

```python
n = int(input("양의 정수를 입력하시오"))

for i in range(1, n+1): # 1~n까지 포함하는 range 객체 생성
  if n % i == 0:        # n을 i로 나눌 때 나머지가 0이면
    print(i)            # 약수이므로 화면에 출력
```

*   실습문제4: 아래 표를 출력하는 프로그램을 작성해보자


| 1 * n | 2 * n | 3 * n | 4 * n | ... | 8 * n |
| ------------ | ------------- | ------------- | ------------- | ------------- | ------------- |
| 1 | 2 | 3 | 4 | ...  | 8  |
| 2 | 4 | 6 | 8 | ...  | 16 |
| 3 | 6 | 9 | 12 | ...  | 24 |
| 4 | 8 | 12 | 16 | ...  | 32 |
| 5 | 10 | 15  | 20 | ...  | 40 |
| ... | ... | ... | ... | ...  | ...  |
| 10 | 20  | 30  | 40  | ...  | 80  |




*   요구사항:
*   - 반복문을 사용한다.
*   - 같은 행에서 셀(cell)은 탭 문자로 분리한다.

```python
for i in range(1, 9):
  if i < 8:
    print(f"{i} * n", end='\t') # 탭 문자로 구분하기
  else:
    print(f"{i} * n")          # 마지막 요소, 줄바꿈 문자 출력해서 행을 바꿈

for n in range(1, 11):
  for i in range(1, 9):
    if i < 8:
      print(i * n, end = '\t')  # 탭 문자로 구분하기
    else:
      print(i * n)
```
