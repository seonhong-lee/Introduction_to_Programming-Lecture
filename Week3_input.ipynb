

*   실습문제8: 커피 원두 100g의 단가와 원두 무게를 사용자로부터 입력받고, 커피 원두 가격을 계산해서 화면에 출력한다.
*   요구사항:
*   1) 커피 원두 100g의 단가는 원 단위의 정수로 입력받고, 무게는 실수로 입력받는다.
*   2) 문자열 출력은 f-문자열을 사용한다.




```python
coffee100 = int(input("커피 원두 100g의 단가를 입력하세요: "))
weight = float(input("구매할 커피 원두의 무게를 입력하세요: "))
print(f"커피 원두 {weight}g 가격: {weight / 100 * coffee100}")
```
